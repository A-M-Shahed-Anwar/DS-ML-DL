### Importing matplotlib.pyplot
pyplot is the visual object of the matplotlib module. <br>
When using notebook please use this magic method ```%matplotlib inline```. This will help us to print plots without the print method like we have printed many things without the print method before.

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

Lets consider two python list: <br>
- x = [1,2,3,4,5,6,7]
- y = [50,51,52,48,47,49,46]

x represents the Days in a week and y represents the temperature on that day.

````python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
````

### plot(x-axis value, y-axis value)
We would like to plot a graph of x and y. We can achieve that by using ```plot()``` method. Visit this [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) for future reference

````python
plt.plot(x,y)
````

As you can see in the plot, the x list which represented the days of a week is in the x-axis. Its because we have given x as the first input and y as the second input ```plot(x,y)```. If we have written ```plot(y,x)``` it would have been the other way around.

#### Color of the graph
In the plot method we can have color option via ```plot(x,y,color='green')```. The default color is blue. After this, the graph color will be green for this graph

````python
plt.plot(x,y,color = 'green')
# There is a shorter version
# plt.plot(x,y,'g')
````

#### linewidth and linestyle
- We can change the linewidth of the graph via ```linwidth=5```
- we can also change the linestyle of the graph via ```linestyle='dotted'```

````python
plt.plot(x,y,color='red',linewidth=5, linestyle = 'dotted')
# # There is shortcut for this
# plt.plot(x,y,'r:', linewidth = 5)
````

#### Labeling (Name the X-axis and Y-axis) - xlabel(), ylabel(), title()
In the figures above we have not marked X-axis and Y-axis. There are just some bunch of numbers. Which will be very hard for our readers. So its the best practice to always name the axes. Also we can give the graph a name with the ```title()``` method. It has a loc attribute, to loc the title in the 'center', 'left' or 'right'

````python
plt.plot(x,y,'y-.')
plt.xlabel('Days of the week')
plt.ylabel('Temperature')
plt.title('Weather Report', loc='left')
````
