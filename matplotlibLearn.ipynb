### Importing matplotlib.pyplot
pyplot is the visual object of the matplotlib module. <br>
When using notebook please use this magic method ```%matplotlib inline```. This will help us to print plots without the print method like we have printed many things without the print method before.

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

Lets consider two python list: <br>
- x = [1,2,3,4,5,6,7]
- y = [50,51,52,48,47,49,46]

x represents the Days in a week and y represents the temperature on that day.

````python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
````

### plot(x-axis value, y-axis value)
We would like to plot a graph of x and y. We can achieve that by using ```plot()``` method. Visit this [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) for future reference

````python
plt.plot(x,y)
````

As you can see in the plot, the x list which represented the days of a week is in the x-axis. Its because we have given x as the first input and y as the second input ```plot(x,y)```. If we have written ```plot(y,x)``` it would have been the other way around.

#### Color of the graph
In the plot method we can have color option via ```plot(x,y,color='green')```. The default color is blue. After this, the graph color will be green for this graph

````python
plt.plot(x,y,color = 'green')
# There is a shorter version
# plt.plot(x,y,'g')
````

#### linewidth and linestyle
- We can change the linewidth of the graph via ```linwidth=5```
- we can also change the linestyle of the graph via ```linestyle='dotted'```

````python
plt.plot(x,y,color='red',linewidth=5, linestyle = 'dotted')
# # There is shortcut for this
# plt.plot(x,y,'r:', linewidth = 5)
````

#### Labeling (Name the X-axis and Y-axis) - xlabel(), ylabel(), title()
In the figures above we have not marked X-axis and Y-axis. There are just some bunch of numbers. Which will be very hard for our readers. So its the best practice to always name the axes. Also we can give the graph a name with the ```title()``` method. It has a loc attribute, to loc the title in the 'center', 'left' or 'right'

````python
plt.plot(x,y,'y-.')
plt.xlabel('Days of the week')
plt.ylabel('Temperature')
plt.title('Weather Report', loc='left')
````

### Format String in the Plot function
We have seen things like ```'g:'``` or ```'y-.'```. Now we will see all the other options there are for matplotlib.

````python
plt.plot(x,y,'rD-') # The order doesn't matter in format string.
````

Now if we don't want to use the format string option, we can specify each attribute. ```color='red'```, ```marker='D'```, ```linestyle='-'```. The output will be the same. **marker** and **linestyle** is a bit tricky to understand. We can also use hexadecimal color code. ```color = '#008840'```. Remember the ```linewidth=5``` attributes we have used before? it will not be in the formatted string.

````python
plt.plot(x,y,color='red',marker='D',linestyle = '-') 
````

There is also other options like ```markersize=20```, ```alpha=1```. markersize to control the size of the marker and alpha is the transparency controller (0 to 1)

````python
plt.plot(x,y,'r+-',markersize = 20, alpha = 0.6)
````

### Axes labels, Legend, Grid

For this section we will increase out data.
- days = [1,2,3,4,5,6,7]
- max_t = [50,51,52,48,47,49,46]
- min_t = [43,42,40,44,33,35,37]
- avg_t = [45,48,48,46,40,42,41]

````python
days = [1,2,3,4,5,6,7] # Weekdays
max_t = [50,51,52,48,47,49,46] # Maximum temp is rocorded on these days
min_t = [43,42,40,44,33,35,37] # Minimum temp is recorded on these days
avg_t = [45,48,48,46,40,42,41] # Average temp is recorded on these days
````

````python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
````

If we look into the above plot, we see bunch of numbers and three lines. Our readers have no idea what does this graph represent? Thats why we will use labelling like we have seen before.

````python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
plt.xlabel('Weekdays')
plt.ylabel('Temperatures')
plt.title('Weather report')
````

Better than before. But still, the graph doesn't make sense. What does these three lines mean? We will use legend to clarify this.

#### Legend()

````python
plt.plot(days, max_t, label = 'Max Temp')
plt.plot(days, min_t, label = 'Min Temp')
plt.plot(days, avg_t, label = 'Avg Temp')
plt.xlabel('Weekdays')
plt.ylabel('Temperatures')
plt.title('Weather report')
````

What the? The labels are not showing up. Because we have not set up legend property yet. ```plt.legend()``` will show the legend property. It has a ```loc='best'``` attribute. The 'best' options looks for the free space on the graph.

````python
plt.plot(days, max_t, label = 'Max Temp')
plt.plot(days, min_t, label = 'Min Temp')
plt.plot(days, avg_t, label = 'Avg Temp')
plt.xlabel('Weekdays')
plt.ylabel('Temperatures')
plt.title('Weather report')
plt.legend()
````

#### Grid()
grid() functions will make the graphs like we used to in the geomatry class

````python
plt.plot(days, max_t, label = 'Max Temp')
plt.plot(days, min_t, label = 'Min Temp')
plt.plot(days, avg_t, label = 'Avg Temp')
plt.xlabel('Weekdays')
plt.ylabel('Temperatures')
plt.title('Weather report')
plt.legend()
plt.grid()
````

So here is the conclusion, two obtain the legend we have to make two steps.
<ol>
<li>label the plot first in the plot(label = 'label name')</li>
<li>activate the legend by plt.legend()</li>
<ol>

#### fill_between()
We can also fill the areas of the below the curves by using `fill_between()` method.

````python
plt.plot(days, max_t, label = 'Max Temp')
plt.fill_between(days, max_t, color='blue', alpha = 0.3)
plt.plot(days, min_t, label = 'Min Temp')
plt.fill_between(days, min_t, color='green', alpha = 0.3)
plt.plot(days, avg_t, label = 'Avg Temp')
plt.fill_between(days, avg_t, color='orange', alpha = 0.3)
plt.xlabel('Weekdays')
plt.ylabel('Temperatures')
plt.title('Weather report')
plt.legend()
plt.grid()
````

### Bar Chart
Use bar chart in data science to visually compare different categories or groups of data, especially when dealing with categorical or nominal data, and for showcasing counts, totals or proportions.<br>
1. **Category Comparison**: When you want to compare values across different categories (sales by product type, population by country, or expenses by department)
2. **Highlighting Differences**: To emphasize the differences between groups or show rankings (e.g., top-selling products or survey responses).
3. **Time Trends (Discrete)**: For displaying changes in data over time when the time intervals are distinct and not continuous (e.g., annual revenue, monthly sales).
4. **Grouped or Stacked Data**: When comparing subcategories within main categories using grouped or stacked bars (e.g. male vs. female participation in events across different years.)
<br>
<br>
![Bar Chart](https://www.w3schools.com/python/img_matplotlib_bars1.png) 


lets say we have the stock market data of some big shot it company. We have tow list: One has the name of the company and another one to represent the revenue of those companies in a particular year.

````python
company = ['Googl', 'Amzn', 'msft', 'fb']
revenue = [90,136,89,27]
````

This data looks like a category comparison where the companies are category and their revenues are being compared. So we can draw a bar chart.

````python
plt.bar(company, revenue, label = 'Revenue')
plt.ylabel('Revenue(bln)')
plt.title('Revenue Report')
plt.legend()
````

This looks good. However now we have the profit data of those companies.

````python
profit = [40, 2, 34, 12]
````

If we want to highlight the differences between the revenue and the profit, we can simpy do so by plotting another barchart.

````python
plt.bar(company, revenue, label = 'Revenue')
plt.bar(company, profit, label = 'profit')
plt.ylabel('Revenue(bln)')
plt.title('Revenue Report')
plt.legend()
````

This looks pretty good in my opinion. However, we can also show the revenue and the profit in a separate bar. We have to use numpy for that. So here is the breakdown on how am I gonna achieve this. The company name in the X-axis is in categorical form, we can convert it into a numpy array.

````python
import numpy as np

xpos = np.arange(len(company)) 
xpos # We have four companies, so we have 4 elements in the array.T
````

Now we are gonna plot this. However you may notice something odd.

````python
plt.bar(xpos, revenue, label = 'Revenue')
plt.bar(xpos, profit, label = 'Profit')
plt.ylabel('Bln $')
plt.title('Us IT company stock')
plt.legend()
````

#### xticks()
Instead of the company name in the X-axis, we have the bars on 0th, 1st, 2nd and 3rd position. We want to bring the company name back and we can achieve that by using xticks() method. 

````python
plt.bar(xpos, revenue, label = 'Revenue')
plt.bar(xpos, profit, label = 'Profit')
plt.xticks(xpos, company)
plt.ylabel('Bln $')
plt.title('Us IT company stock')
plt.legend()
````

We have achieved what we have before except we have imported one library, applied two extra steps in the process. But believe me, now we can achieve something, that was impossible before.
we can separate the revnue and profit bar. The code is self explanatory. **Just remember, the width of the bar must be equal to the change of position**. In our case the change is 4.

````python
plt.bar(xpos-0.2, revenue, width=0.4, label = 'Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label = 'Profit')
plt.xticks(xpos, company)
plt.ylabel('Bln $')
plt.title('Us IT company stock')
plt.legend()
````

We can also achieve horizontal bar chart.

````python
plt.barh(xpos-.2, revenue, height = 0.4, label = 'Revenue')
plt.barh(xpos+.2, profit, height = 0.4,  label = 'Profit')
plt.yticks(xpos, company)
plt.xlabel('Bln $')
plt.title('Us IT company stock')
plt.legend()
````

#### Putting the actual amount on top of the bar

We will use plt.text(x cordinate, y cordinate, text, position) 

- bar.get_x() = Get the starting position of the bar
- bar.get_width()/2 = Get the width of the bar and divide it by 2
- bar.get_height = Get the height of the bar
- s = The string 
- ha = Horizontal Alignment

````python
revenue_bar = plt.bar(xpos-0.2, revenue, width=0.4, label = 'Revenue')
profit_bar = plt.bar(xpos+0.2, profit, width=0.4, label = 'Profit')
plt.xticks(xpos, company)
plt.ylabel('Bln $')
plt.title('Us IT company stock')

for bar in revenue_bar: # To put text at the top of every revenue bar
    x_cor = bar.get_x() + bar.get_width()/2
    y_cor = bar.get_height()
    plt.text(x_cor, y_cor + 2, s = f'{y_cor} bln', ha = 'center')
    
for bar in profit_bar: # To put text at the top of every profit bar
    x_cor = bar.get_x() + bar.get_width()/2
    y_cor = bar.get_height()
    plt.text(x_cor, y_cor + 2, s = f'{y_cor} bln', ha = 'center')
    
plt.legend()
plt.tight_layout()
````

### Histogram

Histograms are best used when you want to visualize the _distribution_ of a dataset, especially for continuous numerical data. They are ideal in these situation:

1. **Data Spread**: To show how values in a dataset are distibuted across different intervals or ranges. For example, the frequency of students achieving scores within scores intervals.
2. **Understanding Patterns**: Useful for identifying patterns like whether the data is skewed, uniform, or normally distributed.
3. **Bin-based Analysis**: When dividing data into bins to analyze frequency counts within specific ranges. For example, Distribute patients frequecny based on Non-dabetic, Pre-diabetic and Diabetic.
4. **Detecting Outliers**: Histograms help pinpoint unusual values or gaps in data distribution.
<br>

Histograms are most suitable for:
- Continuous Data (e.g., height of individuals, temperature readings, or sales amounts.)
- Frequency Data (e.g., The number of occurances within specific ranges.)

***We will use histogram on single column/feature, Because it will count the frequency***

````python
blood_sugar = [113, 95, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

# rwidth = Relative width

plt.hist(blood_sugar, rwidth = .95)
````

We got many outputs, Lets describe them one by one: <br>
The first array (array([3., 2., 2., 0., 1., 1., 0., 2., 0., 2.])) it gives us the frequency of each range/bin/bucket. <br>
The next array (array([ 77. ,  84.3,  91.6,  98.9, 106.2, 113.5, 120.8, 128.1, 135.4, 142.7, 150. ])) It gives us our default 10 segments/bins/bucket. <br>
Here you can see, bin 77-84.3 has a frequency of 3, similarly, bin 98.9 - 106.2 have a frequency of 0, so it is empty in the graph.

````python
blood_sugar = [113, 95, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

# bins = number of bins/buckets in the grapsh (default 10)

plt.hist(blood_sugar, rwidth = .95, bins = 3)
````

We can customize the bin range. 

````python
blood_sugar = [113, 95, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

# bins = number of bins/buckets in the grapsh (default 10) / It can also take an array of range

plt.hist(blood_sugar, rwidth = .95, bins = [80,100,125,150], color = 'g')
plt.title('Blood Sugar Map')
plt.xlabel('Amount of sugar')
plt.ylabel('Number of patients')
````

Now lets say we have two distribution. Blood sugar for men and blood sugar for women. Lets see how we can put two histograms in one graph

````python
blood_sugar_men = [113, 95, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], rwidth = .95, bins = [80,100,125,150], color = ['c', 'b'], label = ['men', 'women'])
plt.title('Blood Sugar Map')
plt.xlabel('Amount of sugar')
plt.ylabel('Number of patients')
plt.legend()
````

Here, we have given the two distributions in an array, tow different **color**s in an array as well as two differnet **label**s in an array. 
> use the color = [c,b]<br>
> use the label = ['men', 'women']<br>
<br><br>
Now Lets say, we want to see it horizontally-
> use the orientation='horizontal' attribute

````python
blood_sugar_men = [113, 95, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], rwidth = .95, bins = [80,100,125,150], color = ['c', 'b'], label = ['men', 'women'], orientation = 'horizontal')
plt.title('Blood Sugar Map')
plt.xlabel('Amount of sugar')
plt.ylabel('Number of patients')
plt.legend()
````

### Pie Chart

When Should we use Pie Chart? <br>
Pie chart are best suited for displaying proportions or percentages in a dataset when you want to show how different parts contribute to a whole. They are most effective when:
1. **You have a small number of categories** (ideally fewer than 6). Too many slices can make the chart cluttered and hard to interpret.
2. **You want to emphasize one part's contribution** compared to others, such as the largest or smallest share.
3. **The total is meaningful** - pie charts work well when the sum of your data adds up to 100%

For example, you might use a pie chart to show:
- Market share of companies in an industry.
- Budget allocation across departments.
- Survey results as percentages (e.g., favorite ice cream flavors).

````python
expenses = [1400, 600, 300, 410, 250]
expense_category = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
````

Here you can see I have expenses of a month based on different categories. Lets assume, this is my whole months expenses.

#### plt.pie()
1. The numerical vlaues `expenses`
2. My categories `labels = expenses_category` 

````python
plt.pie(expenses, labels=expense_category)
plt.show()
````

This pie chart is very small, I want to make it bigger. I will use the `radius` attribute. The default is 1. so if we write `radius = 2`, then the graph will be double the current size. Its way too big. We have settled with 1.5

````python
plt.pie(expenses, labels = expense_category, radius=1.5)
plt.show()
````

````python
# Now we want to bring the percentage. We have to use a attribute called autopct and the string value is '%f%%'.
# Now if we want to show the 2 decimal point we should write '%0.2f%%'
# If we don't want any decimal value, we should write '%0.0f%%'

plt.pie(expenses, labels=expense_category, radius=1.5, autopct='%0.2f%%')
plt.show()
````

````python
# We can also enable shadow. It gives a nice look

plt.pie(expenses, labels=expense_category, radius=1.5, autopct='%0.2f%%', shadow=True)
plt.show()
````

````python
# We can explode slices. You should count slices counter clockwise from the 0 degree angle.
# So, here the first slice should be Home rent, Food, Phone bill, car, other Utilities
# we can explode by providing an array in theis order [0,0.1,0,0.1,0] with the value of explodeness.

plt.pie(expenses, labels=expense_category, radius=1.5, autopct='%0.2f%%', shadow=True, explode= [0, 0.1, 0, 0.1, 0])
plt.show()
````

````python
# We can also change the start angle with the startangle attribute
plt.pie(expenses, labels=expense_category, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.1,0,0.1,0], startangle=180)
plt.show()
````
